```python
hf = h5py.File("SkyNet_r-process.h5")

#This cell is simply for temperature, and since it does not change no matter the elements used, I do not use a function
#At some point, I will have to find a way to find the places to calculate the indices to use to find slope for other hdf5s
temperature = hf["Temperature"]
time = hf["Time"]
mod_temp = []
full_temp = []

linear = np.polyfit(np.log10(np.array(time[900:1700])),np.log10(np.array(temperature[900:1700])),1)
m = linear[0]

for i in range(0,len(temperature)):
    if i < 1700:
        mod_temp.append(np.log10(temperature[i]))
        full_temp.append(temperature[i])
    else:
        del_t = m*(np.log10(time[i]) - np.log10(time[i-1]))
        mod_temp.append(mod_temp[i-1]+del_t)
        full_temp.append(10**mod_temp[i])
        
abundance = hf["Y"]
charge = hf["Z"]
density = hf["Density"]
kbEvpGK = 8.617e-5*1e9
```

```python
#This cell is looking at the differences in the relative abundances between the data sets for lanthanides_actnides and All
#I normalize each data set by the totals, and then I subtract then. Finally, I normalize the differences by using |x|
#However, I just learned that the cutoffs and all the error warnings shown in red are from Ytot being 0 sometimes
#For 5 ionization states of Sm
#This plot is for differences vs time
lan_act = YI_lanth_act[4][:,0:5]
total = YI_all[61][:,0:5]
lanth_tot = np.sum(lan_act,1)
total_tot = np.sum(total,1)

for i in range(len(lanth_tot)):
    lan_act[i,:]/=lanth_tot[i] 
    total[i,:]/=total_tot[i]
    
differences = np.array(total) - np.array(lan_act)
Ytot = np.nansum(np.abs(differences),1)

for i in range(5):
    plt.semilogx(np.array(time),np.array(differences[:,i])/Ytot,label = i)
    #plt.semilogx([ion_potential_Sm[i], ion_potential_Sm[i]], [0, 1],color = 'Black')
    plt.legend()
    plt.xlabel('Time(sec)')
    plt.ylabel('Rel Abundances Differences, Normalized')
    plt.title('Rel Abundance Differences vs Time ')
    plt.show()
```

```python
#This plot is for differences vs time for the 5 ionization state analysis for Sm
for i in range(5):
    plt.semilogx(kbEvpGK*np.array(full_temp),np.array(differences[:,i])/Ytot,label = i)
    plt.semilogx([ion_potential_Sm[i], ion_potential_Sm[i]], [0, 1],color = 'Black')
    plt.legend()
    plt.xlabel('Temp(sec)')
    plt.ylabel('Rel Abundances Differences, Normalized')
    plt.title('Rel Abundance Differences vs Temp ')
    plt.show()
```
