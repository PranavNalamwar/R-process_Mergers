```python
from saha import SkyNetIonization 

fname = 'SkyNet_r-process.h5'

# This is a new class defined in saha.py that reads in all of the 
# SkyNet data and puts it in a form for use by the Saha functions.
# The self-consistent YeFree calculation is done on instantiation 
# if caculateYeFree = True. If calculateYeFree = False, then it is 
# assumed YeFree = YeTotal 
run = SkyNetIonization(fname, calculateYeFree = True)
```

```python
import numpy as np

# Get the isolectronic abundances from zmin=57 to zmax=71
# YI should have shape (times, zmax+1)
YI = run.GetIsoElectronicAbundances(57,71)

YRangeTot = np.sum(YI, 1) # Get the total abundance at each time to normalize

import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
# Plot up a comparison of the fixed YeFree solution and the self-consistent YeFree solution 
for i in range(len(YI[0,:])): 
    plt.plot(run.T9, YI[:,i]/YRangeTot)

plt.yscale('linear')
plt.xscale('log')
plt.xlabel('T (GK)', fontsize=20)
plt.ylabel('Y/Ytot', fontsize=20)
plt.ylim([1.e-6, 1.0])
plt.xlim([1.e-7, 0.5])
plt.show()
np.shape(YI)
```
