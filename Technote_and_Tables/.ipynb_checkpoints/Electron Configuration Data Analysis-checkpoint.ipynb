This document exists to better study the electron configurations of the lanthanides. These results were inputted directly into the Technote.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_excel('electon_configuration abundances_JM_15_09_2020.xls')[0:45]
data
```

```python
data.columns
```

```python
stuff = data[' List of jj configurations with a weight >= 0.01%']
```

```python
array = []
repeats = []

for k in range(len(stuff)):
    
    start = 0
    repeat = 0
    
    for i in range(len(stuff[k])):
        
        if stuff[k][i] == '%':
            
            array.append(stuff[k][start:i])
            start = i
            repeat+=1
            
    repeats.append(repeat)
```

```python
values = []
configs = []

for i in range(len(array)):
    
    temp = str.split(array[i])
    config_temp = ''
    
    for j in range(len(temp)-1):
        
        config_temp+= temp[j] + ' '
    
    configs.append(config_temp)
    values.append(temp[-1])
    
for i in range(len(configs)):
    
    if configs[i][0] == '%':
        configs[i] = configs[i][1:]
```

```python
elements = []
charge_state = []
J = []
total_energy = []
mini_configs = []

for i in range(len(repeats)):
    
    for j in range(repeats[i]):
        elements.append(data['Element'][i])
        charge_state.append(data['Charge State'][i])
        J.append(data['J'][i])
        total_energy.append(data['Total Energy (eV)'][i])
        mini_configs.append(data['Electron Configuration'][i])
```

```python
d = {'Element':elements,'Charge State': charge_state,'Orbitals(NIST)': mini_configs,'Configuration': configs,'J':J,'Total Energy':total_energy,'Probabilities':values}
pd.DataFrame(d).to_csv('Configuration_Table.csv')
table = pd.DataFrame(d)
table.iloc[[2,15],:]
```

```python
shared_nonrestrict = dict()
shared_restrict = dict()

for i in range(len(table['Configuration'])):
    indices = []
    
    for j in range(len(table['Configuration'])):
        
        if i!=j and table['Configuration'][i] == table['Configuration'][j] and table['Element'][i] != table['Element'][j]:
            indices.append(j)
            
    shared_restrict[i] = indices
    
for i in range(len(table['Configuration'])):
    indices = []
    
    for j in range(len(table['Configuration'])):
        
        if i!=j and table['Configuration'][i] == table['Configuration'][j]:
            indices.append(j)
            
    shared_nonrestrict[i] = indices
    
shared_nonrestrict
```
